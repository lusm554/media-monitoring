# Data

```python
import sys
import pandas as pd
sys.version, pd.__version__
```

```python
raw_news = pd.read_csv('raw_news_samples.csv')
raw_news = raw_news[[c for c in raw_news.columns if not 'Unnamed:' in c]]
raw_news.info()
```

```python
raw_news.shape
```

```python
raw_news[['title', 'body_text']] = raw_news[['title', 'body_text']].fillna('')
```

```python
raw_news.scraper.value_counts()
```

```python
raw_news[['title', 'body_text']].map(len).agg(['mean', 'max', 'min', 'median']).round().T
```

# Разметка заголовков новостей

```python
target_news = raw_news[['url', 'title']]
target_news = pd.concat([pd.read_csv('google_news.csv')[['url', 'title']], target_news])
target_news.info()
```

```python
target_news['target'] = target_news.title.apply(lambda x: int('цфа' in x.lower()))
```

```python
target_news['target'].value_counts()
```
